## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
from scipy.stats import linregress
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_complete = pd.merge(mouse_metadata, study_results, how = "left", on = "Mouse ID")

# Display the data table for preview
data_complete.head()
```

```python
# Checking the number of mice.
total_mice = len(data_complete["Mouse ID"].unique())
print(total_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_mice = data_complete.loc[data_complete.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
dup_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.
dup_mice_df = data_complete.loc[data_complete["Mouse ID"] == "g989", :]
dup_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mice_df = data_complete[data_complete['Mouse ID'].isin(dup_mice)==False]
clean_mice_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mice = len(clean_mice_df["Mouse ID"].unique())
print(mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean = group_df = clean_mice_df['Tumor Volume (mm3)'].groupby(clean_mice_df["Drug Regimen"]).mean()
median = group_df = clean_mice_df['Tumor Volume (mm3)'].groupby(clean_mice_df["Drug Regimen"]).median()
variance = group_df = clean_mice_df['Tumor Volume (mm3)'].groupby(clean_mice_df["Drug Regimen"]).var()
sta_dev = group_df = clean_mice_df['Tumor Volume (mm3)'].groupby(clean_mice_df["Drug Regimen"]).std()
SEM = group_df = clean_mice_df['Tumor Volume (mm3)'].groupby(clean_mice_df["Drug Regimen"]).sem()

# Assemble the resulting series into a single summary dataframe.
summary = pd.DataFrame({"Mean":mean, "Mediam":median, "Variance":variance, "Standard Deviation":sta_dev, "SEM":SEM})
summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
summary_agg = clean_mice_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
drugs = data_complete[data_complete["Drug Regimen"].isin(drug_list)]

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = drugs.groupby(["Drug Regimen", "Mouse ID"]).agg(tumor_size = ("Tumor Volume (mm3)", lambda x: x.iloc[-1]))

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_volume = pd.merge(data_complete, last_timepoint, how = "left", on = "Mouse ID")
tumor_volume.head()
```

```python
last_timepoint = last_timepoint.stack(level=0).unstack(level=0)
for drug in drug_list:
    print(drug)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset
    # Determine outliers using upper and lower bounds
boxplot_list = []
for drug in drug_list:
    quartiles = last_timepoint[drug].quantile([.25,.5,.75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    boxplot_list.append(list(last_timepoint[drug].dropna()))
    print(f"------------------------------------------------------------")
    print(f"The lower quartile of {drug} treatments is: {lowerq}")
    print(f"The upper quartile of {drug} treatments is: {upperq}")
    print(f"The interquartile range of {drug} treatments is: {iqr}")
    print(f"Values below {lower_bound} could be {drug} outliers.")
    print(f"Values above {upper_bound} could be {drug} outliers.")
    print(f"------------------------------------------------------------")
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.figure(5)
plt.boxplot(boxplot_list)
plt.xlabel("Regimen")
plt.xticks([1,2,3,4], drugs_list, rotation=90)
plt.ylabel("Tumor Volume")
plt.title("Tumor Volume by Drug Regimen")
plt.tight_layout()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
plt.figure(6)
Capomulin_df = clean_mice_df.loc[clean_mice_df["Drug Regimen"] == "Capomulin",:]
forline_df = Capomulin_df.loc[Capomulin_df["Mouse ID"] == 's185']

x_axis = forline_df["Timepoint"]
tumor_size = forline_df["Tumor Volume (mm3)"]
plt.title('Capomulin treatmeant of mouse s185')
plt.plot(x_axis, tumor_size,linewidth=2, markersize=5,marker="*",color="purple")
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
plt.figure(7)
average_cap_volume =Capomulin_df.groupby(['Mouse ID']).mean()
plt.scatter(average_cap_volume['Weight (g)'], average_cap_volume['Tumor Volume (mm3)'], marker="*", color="purple")
plt.title("Weight vs Tumor Volume for Capomulin")
plt.xlabel('Weight')
plt.ylabel('Average Tumor Volume')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
(slope, intercept,rvalue, pvalue, stderr) = linregress(average_cap_volume["Weight (g)"],average_cap_volume["Tumor Volume (mm3)"])
regress_values=average_cap_volume["Weight (g)"]* slope + intercept
line_eq= f"y = {round(slope, 2)} x + {round(intercept, 2)}"

plt.scatter(average_cap_volume["Weight (g)"],average_cap_volume["Tumor Volume (mm3)"],color='purple')
plt.plot(average_cap_volume["Weight (g)"], regress_values, color='purple')
plt.annotate(line_eq,(20,36), fontsize=14)
plt.xlabel("Weight")
plt.ylabel("Average Tumor Volume")
plt.title("Weight vs Tumor Volume for Capomulin")
```

```python
correlation = st.pearsonr(average_cap_volume['Weight (g)'],average_cap_volume['Tumor Volume (mm3)'])
print(f"The correlation is {round(correlation[0],2)}")
print(f"The r-squared is: {round(rvalue**2,3)}")
```
